```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt",skiprows=1)
```

```python
print(data.shape) #there are 2642 rows and 2 cols
magnitude = data[:,0]
temperature = data[:,1]
print(np.min(temperature))
print(np.max(temperature))
print(np.min(magnitude)) #Schedar
print(np.max(magnitude))
```

```python
logmag = np.log(magnitude)
logtemp = np.log(temperature)
plt.figure(figsize=(8, 6))
plt.scatter(logtemp,logmag)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.xlabel("logtemp")
plt.ylabel("logmag")
brightidx = np.argmin(logmag)
dimidx = np.argmax(logmag)
hotidx = np.argmax(logtemp)
coldidx = np.argmin(logtemp)
plt.scatter(logtemp[brightidx],logmag[brightidx],color="black",marker='*', s=200, label="brightest")
plt.scatter(logtemp[dimidx],logmag[dimidx],color="purple",marker='*', s=200, label="dimmest")
plt.scatter(logtemp[hotidx],logmag[hotidx],color="green",marker='*', s=200, label="hottest")
plt.scatter(logtemp[coldidx],logmag[coldidx],color="orange",marker='*', s=200, label="Coldest star")
plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.legend().set_zorder(1)
```
