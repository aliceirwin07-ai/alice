```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def makesinwave(x, A, w):
    return A*np.sin(w*x)

x = np.linspace(0,2*np.pi,1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
fig, axes = plt.subplots(1,3,figsize=(25,5))
axes[0].set_title("amplitude",fontsize=16)
axes[1].set_title("frequency",fontsize=16)
axes[2].set_title("amplitude & frequency",fontsize=16)
axes[0].set_xlabel("x",fontsize=16)
axes[1].set_xlabel("x",fontsize=16)
axes[2].set_xlabel("x",fontsize=16)
axes[0].set_ylabel("sin(x)",fontsize=16)
axes[1].set_ylabel("sin(x)",fontsize=16)
axes[2].set_ylabel("sin(x)",fontsize=16)
for amp,freq in zip(amplitude,frequency):
    axes[0].plot(x,makesinwave(x,amp,1))
    axes[1].plot(x,makesinwave(x,1,freq))
    axes[2].plot(x,makesinwave(x,amp,freq))
plt.show()
```
